```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 1. Load the dataset
# (Make sure "Linear.csv" is in your working directory)
Linear = pd.read_csv("Regression_problem.csv")
Linear
```

```python
 #2. Explore the data
print(Linear.head())
print(Linear.info())
print(Linear.describe())
print("Columns:", Linear.columns)

```

```python
# 3. Define Features (X) and Target (y)
# Note: The column names in your CSV seem to have leading spaces (e.g., ' Size')
x = Linear[['example','Size','Place']]
y = Linear['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3,random_state=42)
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.intercept_
lr.coef_
y_pred = lr.predict(x_test)
print("Predicted Prices:", y_pred)
print("Actual Prices:", y_test.values)


plt.scatter(y_test,y_pred)
plt.plot(y_test, y_pred, 'g-')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.show()
```
